```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing
```

```python
company = pd.read_csv('Company_Data (1).csv')
```

```python
company
```

```python
result = []
for value in company["Sales"]:
    if value <= 8 :
        result.append("Bad")
    else :
        result.append("Good")
    
company["Sales"] = result
print(company)
```

```python
company.Sales.value_counts()
```

```python
labelenc = LabelEncoder()
```

```python
company['Sales']= labelenc.fit_transform(company['Sales']) 
company['US']= labelenc.fit_transform(company['US'])
company['ShelveLoc']= labelenc.fit_transform(company['ShelveLoc']) 
company['Urban']= labelenc.fit_transform(company['Urban']) 
```

```python
company
```

```python
x = company.iloc[:,1:]
y = company.iloc[:,0:1]
```

```python
x
```

```python
y
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Performing Random Forest classifier

num_trees = 100 
max_features = 3
kfold = KFold(n_splits=5)
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
results = cross_val_score(model, x, y, cv=kfold)
print(results.mean())
```
