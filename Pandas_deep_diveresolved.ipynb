## 4. create a function called major that returns a boolean value to a new column ' legal drinker'.
* Note. the age eligibility is 17

```python
student_legal_drinker = student_data.loc[:,'school':'guardian']
student_legal_drinker
```

```python
student_legal_drinker['legal_drinker'] = student_legal_drinker.age.apply(lambda x : x >= 17)
student_legal_drinker
```

```python
def major(x):
    if x>17:
        return True
    else:
        return False
student_legal_drinker['legal_drinkers'] = student_legal_drinker.age.apply(major)
student_legal_drinker.head(20)
```

## 1. Get the regionwise average insurance charges

## 2. Get the region and sex wise average insurance charges

```python
round(insurance_data.groupby(['region','sex'])['charges'].mean())
```

## Understanding Pivot table

```python
region_wise_insurance_charges = insurance_data[['region','sex','charges','children']]
round(pd.pivot_table(region_wise_insurance_charges , values='charges', index='children', columns='region',aggfunc='mean'))
```

## Understanding crosstab
* It explains the frequency of occurrence of the categorical data we give as index 

```python
pd.crosstab(index= insurance_data.region , columns= insurance_data.sex, margins=True )
```

```python
pd.crosstab(index=insurance_data.region, columns=insurance_data.children, margins=True)
```
